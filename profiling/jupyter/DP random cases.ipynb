```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/dp_random_cases.csv')
```

```python
df = df.rename(index=str, columns={"instance_index": "Instancia", "algorithm_time": "Tiempo"})

plt.clf()
axis1 = plt.subplot(2,1,1)
axis2 = plt.subplot(2,1,2)
df.plot(x='Instancia', y='Tiempo', kind='box', ax=axis1)
df.plot(x='Instancia', y='Tiempo', kind='density', ax=axis2)
plt.show()
```
