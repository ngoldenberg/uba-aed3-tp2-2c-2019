```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/cost_variation_low_nk.csv')
```

```python
plt.clf()
axis1 = plt.subplot(2,1,1)
df.plot(x='iteration', y='time_1', kind='density', ax=axis1)
axis1.legend(['Costos aleatorios'])
axis1.set_ylabel('Densidad')
axis1.set_xlabel('Tiempo')
plt.show()
```
