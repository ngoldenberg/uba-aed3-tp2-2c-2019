```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy import stats
```

```python
df = pd.read_csv('../results/cross_over_w.csv')
```

```python
plt.clf()
fig, ax = plt.subplots()

for key, grp in df.groupby(['algorithm']):
    ax = grp.plot(ax=ax, kind='line', x='W', y='algorithm_time', label='algorithm = ' + str(key))
    ax.set_ylabel('Tiempo(s)')
    
plt.show()
```

```python
df2 = df.loc[df['W'] == 200]

a = sns.jointplot(np.array([i for i in range(0,1000)]), df2['algorithm_time'], kind="reg").set_axis_labels("x", "tiempo")
a.annotate(stats.pearsonr)

df3 = df.loc[df['W'] == 600]
b = sns.jointplot(np.array([i for i in range(0,1000)]), df3['algorithm_time'], kind="reg").set_axis_labels("x", "tiempo")
b.annotate(stats.pearsonr)

df4 = df.loc[df['W'] == 1000]
b = sns.jointplot(np.array([i for i in range(0,1000)]), df4['algorithm_time'], kind="reg").set_axis_labels("x", "tiempo")
b.annotate(stats.pearsonr)

plt.show()
```
