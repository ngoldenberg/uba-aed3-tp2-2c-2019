```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/k_growth.csv')
```

```python
plt.clf()
df2 = df.drop('k^2', 1)
ax = df2.plot(x='k', logy=True)
ax.legend(['Crecimiento de K'])
ax.set_ylabel('segundos (log)')
ax.set_xlabel('k')
plt.show()
```

```python
sns.jointplot(df['k^2'], df['time'], kind="reg").set_axis_labels("x^2", "tiempo")
sns.plt.show()
```
