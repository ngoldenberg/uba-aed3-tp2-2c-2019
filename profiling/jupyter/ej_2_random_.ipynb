```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/arbitraje_random_cases.csv')
```

```python
plt.clf()
fig, ax = plt.subplots()

for key, grp in df.groupby(['algorithm']):
    ax = grp.plot(ax=ax, kind='line', x='instance_index', y='algorithm_time', label='algorithm = ' + str(key))
    ax.set_ylabel('Tiempo(s)')
    
plt.show()
```
