```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('../results/n_growth.csv')
```

```python
plt.clf()
df2 = df.drop('x^2', 1)
ax = df2.plot(x='n', logy=True)
ax.legend(['Crecimiento de N'])
ax.set_ylabel('segundos (log)')
ax.set_xlabel('n')
plt.show()
```

```python
sns.jointplot(df['x^2'], df['time'], kind="reg").set_axis_labels("x^2", "tiempo")
sns.plt.show()
```
